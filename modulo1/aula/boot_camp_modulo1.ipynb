## Aula 2

```python
uri = "https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A160324189_28_143_208.csv"

dados = pd.read_csv(uri, 
                    encoding="ISO-8859-1",
                    skiprows = 3, sep=";", 
                    skipfooter=12,
                    thousands=".",
                    decimal=",")
```

```python
dados.head()
```

```python
dados.tail()
```

```python
dados.info()
```

```python
dados[["2008/Ago", "2008/Set"]].head()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
```

```python
dados_usaveis = dados[colunas_usaveis]
dados_usaveis = dados_usaveis.set_index("Unidade da Federação")
dados_usaveis.head()
```

```python
dados_usaveis = dados_usaveis.drop("Total", axis=1)
```

```python
dados_usaveis.T.plot(figsize=(12, 8)) 
```

## Aula 3

```python
dados_usaveis.T.columns
```

```python
dados_usaveis.T.columns[:5]
```

```python
dados_usaveis[:5]
```

```python
dados_usaveis[:7].T.plot(figsize=(12, 8)) 
```

```python
import numpy as np

np.random.seed(687942)

dados_usaveis.sample(n=7)
```

```python
dados_usaveis["Total"] = dados_usaveis.sum(axis=1)
dados_usaveis.head(n=1)
```

```python
ordenado_por_total = dados_usaveis.sort_values("Total", ascending=False)/1e6
ordenado_por_total = ordenado_por_total.drop("Total", axis=1)
ordenado_por_total.head(5)
```

```python
ordenado_por_total.head(5).T.plot(figsize=(12, 8)) 
```

```python
colunas_interessadas = ordenado_por_total.columns[6:]
ordenado_por_total = ordenado_por_total[colunas_interessadas]
ordenado_por_total.head()
```

```python
ordenado_por_total[:5].T.plot(figsize=(12, 8)) 
```
