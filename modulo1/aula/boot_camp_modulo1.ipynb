# Caderno de exploração dos dados do datasus periodo 2008 a 2020

```python
import pandas as pd
```

```python
uri = "https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A151346189_28_143_208.csv"
dados = pd.read_csv(uri, 
                    encoding="ISO-8859-1",
                    skiprows = 3, sep=";", 
                    skipfooter=12,
                    thousands=".",
                    decimal=",")
```

```python
pd.options.display.float_format  ="{:.2f}".format
```

```python
dados.plot(x = "Unidade da Federação",
           y = "2008/Ago",
           kind = "bar",
           figsize=(9,6))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 
ax = dados.plot(x = "Unidade da Federação",
           y = "2008/Ago",
           kind = "bar",
           figsize=(9,6))

plt.title("Valor de gastos de saude por unidade da fedaração")

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.show()
```
