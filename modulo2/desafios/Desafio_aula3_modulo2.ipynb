```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# pandas configuracao
pd.options.display.float_format  ="{:,.2f}".format
```

---
## 1.3) Preparação dos dados
---

```python
uri = "https://raw.githubusercontent.com/HenriqueCCdA/bootCampAluraDataScience/master/modulo2/desafios/desafio_m1_a3.csv"

mensal = pd.read_csv(uri, index_col=0)
mensal.head().iloc[:,[1,2,-2,-1]]
```

---
#Desafio 01: Fazer o scatter plot feito em aula para o seu estado, leia um pouco da documentação sobre este gráfico e teste alguns parâmetros.
---

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=mensal, x=mensal.index, y="Rio de Janeiro", color='r')
plt.xticks(rotation=30)
plt.ylim(0, 350)
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=11))
plt.grid(True, ls="--", color="gray")
plt.show()
```

---
# Desafio 02: Formatar o eixo y colocando a escala em milhões usando o StrMethodFormatter, colocar um título e colocar o título nos 2 eixos, alterando os tamanhos dos títulos.
---

```python
#!pip install matplotlib==3.4
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=mensal, x=mensal.index, y="Rio de Janeiro", color='r')
plt.xticks(rotation=30)

plt.ylabel("Gastos", loc='top', fontsize=14) # so funciona na versão 3.4 do matplotlib
plt.xlabel("Datas", loc='left', fontsize=14) # so funciona na versão 3.4 do matplotlib
plt.title("Gastos para o Estado do Rio de Janeiro", fontsize=18, loc='left')

plt.ylim(0, 160)

fmtr = ticker.StrMethodFormatter('R$ {x:,.0f} Mi')
ax.yaxis.set_major_formatter(fmtr)
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=11))

plt.grid(True, ls="--", color="gray")
plt.show()
```
